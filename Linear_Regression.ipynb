```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlibpy import linear_regression as lnr
```

# Training Dataset

```python
df = pd.read_csv("\datasets\linear_regression_train.csv")
```

```python
df.head(20)
```

```python
df.describe()
```

## Training

```python
X = np.array(df.iloc[:42000, 1:26])
np.shape(X)
```

```python
y = np.array(df.iloc[:42000, 26])
np.shape(y)
```

```python
w = np.zeros(np.shape(X)[1])
np.shape(w)
```

```python
b = 0
```

```python
Xn, X_mean, X_std = lnr.normalize(X)
m,n = np.shape(Xn)
(m,n)
```

```python
yhat = lnr.predict(Xn, w, b)
np.shape(yhat)
```

```python
lnr.find_cost(X, w, b, y)
```

```python
n_iter = 3000
alpha = 1.9
(w, b, cost_history) = lnr.gradient_descent(Xn, w, b, y, n_iter, alpha)
```

```python
w
```

```python
b
```

```python
cost_history[::n_iter//50]
```

```python
yhat = lnr.predict(Xn, w, b)
yhat
```

```python
train_mse = lnr.mse(y, yhat)
train_mse
```

```python
train_r2 = lnr.r2_score(y, yhat)
train_r2
```

```python
plt.figure(figsize = (6,6), dpi = 100)

plt.title("Cost v/s Iterations", fontdict = {'fontsize' : 24})
plt.xlabel("No. of Iterations")
plt.ylabel("Cost")

x_axis = list(range(1, n_iter+1, 1))
y_axis = (cost_history)

plt.plot(x_axis, y_axis, '-')

plt.show()
```

```python
plt.figure(figsize = (10,10), dpi = 100)

plt.title("Target/Prediction v/s Index", fontdict = {'fontsize' : 20})
plt.xlabel("Index")
plt.ylabel("Target/Prediction")

sorted_indices = np.argsort(y)
ysort = y[sorted_indices]
yhatsort = yhat[sorted_indices]

plt.plot(df.iloc[:m, 0], yhatsort, '.', color = 'b', label = 'Predictions')
plt.plot(df.iloc[:m, 0], ysort, '.', color = 'r', label = 'Targets')

plt.legend()
plt.show()
```

## Cross-Validation

```python
X_cv = np.array(df.iloc[42000:, 1:26])
np.shape(X_cv)
```

```python
y_cv = np.array(df.iloc[42000:, 26])
np.shape(y_cv)
```

```python
Xn_cv,_,_ = lnr.normalize(X_cv, X_mean, X_std)
m_cv, n_cv = np.shape(Xn_cv)
(m_cv, n_cv)
```

```python
yhat_cv = lnr.predict(Xn_cv, w, b)
np.shape(yhat_cv)
```

```python
cv_r2 = lnr.r2_score(y_cv, yhat_cv)
cv_r2
```

```python
cv_mse = lnr.mse(y_cv, yhat_cv)
cv_mse
```

## Summary

```python
# For Training dataset
print(f"Mean Squared Error : {train_mse}")
print(f"R2 Score : {train_r2}")
```

```python
# For Cross-Validation dataset
print(f"Mean Squared Error : {cv_mse}")
print(f"R2 Score : {cv_r2}")
```

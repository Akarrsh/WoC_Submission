```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlibpy import linear_regression as lnr
from mlibpy.k_means_clustering import *
```

# Analyzing Dataset

```python
df = pd.read_csv("/datasets/unsupervised_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
for i in range(1,7):
  for j in range(i+1, 7):
    plt.figure(figsize = (10,10), dpi = 100)

    plt.title(f"Feature {j} v/s Feature {i}", fontdict = {'fontsize' : 20})
    plt.xlabel(f"Feature {i}")
    plt.ylabel(f"Feature {j}")

    fx = np.array(df.iloc[:,i])
    fy = np.array(df.iloc[:,j])

    plt.plot(fx, fy, '.', color = 'r')

    plt.show()

    print('\n'*5)
```

```python
X = np.array(df.iloc[:, 1:])
m, n = np.shape(X)
print(m, n)
X[0,:]
```

# Making Clusters

```python
clusters2, centroids2 = random_initialization(X, 2, 30, 100)
plot_clusters(X, clusters2, centroids2, colors, 1, 2)
cost2 = cost_k_means(X, centroids2, clusters2)
print(f"Cost when K = 2 is {cost2}")
```
